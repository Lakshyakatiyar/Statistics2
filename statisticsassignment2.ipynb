1=The three main measures of central tendency are:

Mean: The mean is the average of all the data points. It is calculated by adding up all the values and dividing by the total number of values. The mean is sensitive to extreme values and provides a balanced summary when the data is approximately symmetric.

Median: The median is the middle value in a dataset when the values are arranged in ascending order. If there is an even number of values, the median is the average of the two middle values. The median is robust to outliers and extreme values and is particularly useful when the data contains skewed distributions or outliers.

Mode: The mode is the value that appears most frequently in a dataset. It represents the most common value(s) in the data and is particularly relevant for categorical or discrete data.

2=Mean:

The mean is the sum of all values in a dataset divided by the total number of values.
It is sensitive to extreme values and outliers.
The mean is commonly used when the data is approximately symmetric and does not have extreme values that significantly affect the average.
Formula: Mean = (Sum of all values) / (Total number of values)
Median:

The median is the middle value in a dataset when the values are arranged in ascending order.
If there is an even number of values, the median is the average of the two middle values.
The median is robust to outliers and extreme values, making it useful when the data is skewed or contains outliers.
Formula (Odd number of values): Median = Middle value
Formula (Even number of values): Median = (Middle value + Next middle value) / 2
Mode:

The mode is the value that appears most frequently in a dataset.
There can be no mode (no value appears more than once) or multiple modes (multiple values have the highest frequency).
The mode is particularly useful for categorical or discrete data.
It can be used alongside other measures of central tendency, especially when dealing with categorical data.

Measures of dispersion, including range, variance, and standard deviation, are used to quantify the spread or variability of data points within a dataset. They provide insights into how closely or widely data points are distributed around the measures of central tendency (such as the mean or median). Let's explore each measure of dispersion and provide an example:

Range:
The range is the simplest measure of dispersion. It calculates the difference between the maximum and minimum values in a dataset. It gives a rough idea of the spread but is sensitive to outliers.

Example:
Consider the heights (in centimeters) of a group of students: [165, 168, 170, 172, 175, 178, 185].
Range = Maximum height - Minimum height
Range = 185 - 165 = 20 cm
In this example, the range indicates that the heights vary by 20 cm.

Variance:
The variance measures the average of the squared differences between each data point and the mean. It provides a more precise measure of dispersion than the range and considers the entire dataset.

Example:
Consider the test scores of a class: [85, 90, 88, 92, 87].

Calculate the mean: (85 + 90 + 88 + 92 + 87) / 5 = 88.4
Calculate the squared differences from the mean: [(85 - 88.4)^2, (90 - 88.4)^2, ...]
Calculate the mean of squared differences: (20.16 + 4.84 + 0.36 + 13.24 + 0.04) / 5 = 7.336
Variance ≈ 7.336
Standard Deviation:
The standard deviation is the square root of the variance. It provides a measure of dispersion that is in the same units as the original data.

Example (continued from variance example):
Calculate the square root of the variance to find the standard deviation:
Standard Deviation = √Variance
Standard Deviation ≈ √7.336 ≈ 2.710

In this example, the standard deviation of approximately 2.710 indicates the average amount of deviation of test scores from their mean. A higher standard deviation suggests greater variability, while a lower standard deviation suggests less variability.

8.Skewness:

Skewness is a measure of the asymmetry of the probability distribution of a real-valu
e random variable about its mean.

Types of Skewness:
Positive Skewness (Right Skewed): When the tail on the right side of the distribution is longer or fatter than the left side. This means the mean is typically greater than the median.
Negative Skewness (Left Skewed): When the tail on the left side of the distribution is longer or fatter than the right side. This means the mean is typically less than the median.
Skewness helps to understand the shape of the distribution and the direction in which most of the data is spread out.

9.In a right-skewed (positively skewed) distribution, the mean is typically greater than the median. This is because the long tail on the right side pulls the mean to the right more than the median.
